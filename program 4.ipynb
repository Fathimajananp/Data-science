```python
# simple graph

import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4])
y=np.array([1,2,3,4])
plt.plot(x,y,marker='o',linestyle='--',color='g',ms=10,mec='r',mfc='g')
plt.xlabel('width')
plt.ylabel('length')
plt.title('Graph',loc='center')
plt.grid()
plt.show()
```

```python
#bar graph

import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4])
y=np.array([1,2,3,4])
plt.bar(x,y)
plt.xlabel('width')
plt.ylabel('length')
plt.title('Graph',loc='center')
plt.grid()
plt.show()
```

```python
#horizontal bar graph

import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4])
y=np.array([1,2,3,4])
plt.barh(x,y)
plt.xlabel('width')
plt.ylabel('length')
plt.title('Graph',loc='center')
plt.grid()
plt.show()
```

```python
#scatter graph

import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4])
y=np.array([1,2,3,4])
plt.scatter(x,y)
plt.xlabel('width')
plt.ylabel('length')
plt.title('Graph',loc='center')
plt.grid()
plt.show()
```

```python
#histogram graph

import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4])
y=np.array([1,2,3,4])

h=np.random.normal(170,10,250)
plt.hist(h)
plt.xlabel('width')
plt.ylabel('length')
plt.title('Graph',loc='center')
plt.grid()
plt.show()
```

```python
#subplot 

import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4])
y=np.array([1,2,3,4])
plt.subplot(1,2,1)
plt.plot(x,y)
plt.xlabel('width')
plt.ylabel('length')
plt.title('Graph',loc='center')
plt.grid()


plt.subplot(1,2,2)
x=np.array([1,2,3,4])
y=np.array([1,2,3,4])
plt.bar(x,y)
plt.xlabel('width')
plt.ylabel('length')
plt.title('Graph',loc='center')
plt.grid()

plt.show()
```

```python
#piechart

import matplotlib.pyplot as plt
import numpy as np
x=np.array([30,25,20,10])
a=["red","green","yellow","blue"]
b=["red","green","yellow","blue"]
c=[0,0.2,0,0]
plt.pie(x,labels=a,explode=c,shadow=True,colors=b)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([30,25,20,10,5])
a=["red","green","yellow","blue","magenta"]
b=["red","green","yellow","blue","magenta"]
c=[0,0,0.1,0,0]
plt.pie(x,labels=a,explode=c,shadow=True,colors=b)
plt.show()
```
