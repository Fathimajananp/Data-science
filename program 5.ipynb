```python
#rotation

from PIL import Image
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt
%matplotlib inline
ph=Image.open('a.jpg')
b=np.array(ph)
c=ndimage.rotate(b,45)
d=Image.fromarray(c)
d.save('photo1.jpg')
plt.imshow(d)
```

```python
#scaling

from PIL import Image
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt
%matplotlib inline
ph=Image.open('a.jpg')
b=np.array(ph)
c=ndimage.zoom(b,(.25,1,1))
d=Image.fromarray(c)
d.save('photo1.jpg')
plt.imshow(d)
```
